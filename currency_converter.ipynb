```python
# Install Gradio
!pip install gradio requests


# Import necessary libraries
import gradio as gr
import requests

# Function to fetch conversion rate and perform conversion
def convert_currency(amount, from_currency, to_currency):
    if from_currency == to_currency:
        return f"{amount:.2f} {to_currency}"

    try:
        response = requests.get(f"https://api.exchangerate-api.com/v4/latest/{from_currency}")
        response.raise_for_status()
        data = response.json()

        if to_currency in data['rates']:
            conversion_rate = data['rates'][to_currency]
            converted_amount = amount * conversion_rate
            return f"{converted_amount:.2f} {to_currency}"
        else:
            return f"Currency {to_currency} not available"
    except requests.exceptions.RequestException as e:
        return f"Error fetching conversion rates: {e}"

# Define Gradio interface
def currency_converter_interface(amount, from_currency, to_currency):
    try:
        amount = float(amount)
        return convert_currency(amount, from_currency, to_currency)
    except ValueError:
        return "Invalid input. Please enter a numeric value."

# Create the Gradio app
with gr.Blocks() as app:
    gr.Markdown("# Currency Converter")
    amount = gr.Number(label="Amount")
    from_currency = gr.Dropdown(choices=["USD", "EUR", "GBP", "JPY", "INR", "CAD", "AUD"], label="From Currency", value="USD")
    to_currency = gr.Dropdown(choices=["USD", "EUR", "GBP", "JPY", "INR", "CAD", "AUD"], label="To Currency", value="EUR")
    result = gr.Textbox(label="Converted Amount")

    convert_button = gr.Button("Convert")
    convert_button.click(
        fn=currency_converter_interface,
        inputs=[amount, from_currency, to_currency],
        outputs=result
    )

app.launch()
```
